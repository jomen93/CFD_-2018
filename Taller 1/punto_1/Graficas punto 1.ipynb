```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy as sp
%matplotlib inline
```

# Primera derivada

```python
d1 = np.loadtxt('d1_2do.dat', unpack = True)
d2 = np.loadtxt('d1_4to.dat', unpack = True)

plt.loglog(d1[0],d1[1],"r.",label="$2^{do}$ orden")
plt.loglog(d2[0],d2[1],"b.",label="$4^{to}$ orden")

plt.grid(True)
plt.xlabel("$\\Delta.x$")
plt.ylabel("$L_{2}$")
plt.title("primera derivada")
plt.legend()
plt.savefig("1_derivada_df.eps",format = "eps", dpi =400)
```

# Segunda derivada

```python
d3 = np.loadtxt('d2_4to.dat', unpack = True)
d4 = np.loadtxt('d2_6to.dat', unpack = True)

plt.loglog(d3[0],d3[1],"r.",label="$2^{do}$ orden")
plt.loglog(d4[0],d4[1],"b.",label="$4^{to}$ orden")

plt.grid(True)
plt.xlabel("$\\Delta.x$")
plt.ylabel("$L_{2}$")
plt.title("Segunda derivada")
plt.legend()
plt.savefig("2_derivada_df.eps",format = "eps", dpi =400)
```
