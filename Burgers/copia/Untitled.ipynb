```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np



%matplotlib inline
```

```python
u = np.loadtxt("t1uCD.plt")
np.shape(u)
```

```python
U = u[:,2].reshape((26,26))
x = u[:,0][0:26]
y = u[:,1][0:26]
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

X, Y = np.meshgrid(u[:,0][0:21], u[:,1][0:21])
surf = ax.plot_surface(X, Y,U, cmap="jet")#,linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_ylim(0.0, 1.01)
ax.view_init(elev=0, azim=90)
#ax.zaxis.set_major_locator(LinearLocator(10))
#ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
```

```python
plt.contourf(x,y,U,cmap=cm.jet)
#plt.colorbar()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm
import numpy as np

data = np.loadtxt("t2uCD.plt")
#X, Y =np.meshgrid(zip(*data)[axe_labels.index(var_axe1)],zip(*data)[axe_labels.index(var_axe2)])
X,Y=zip(*data)[0],zip(*data)[1]
Z = zip(*data)[2]

#print "max Z: ", max(Z)
fig = plt.figure(figsize=(10,7))
ax = fig.gca(projection='3d')
plt.hold(True)
#surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
surf = ax.plot_trisurf(X, Y, Z, cmap="jet", linewidth=0.2)
cbar = fig.colorbar(surf, shrink=0.7, aspect=8)
cbar.set_label('Magnitud de la velocidad $u$]')
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel('$u$')
ax.view_init(elev=50, azim=10)
ax.set_title("Perfil de velocidad",fontsize = 20,y=1.08)
#ax.view_init(elev=10, azim=-45)
#ax.scatter(X,Y,Z, marker='.', s=10, c="black", alpha=0.5)
#plt.savefig("Densidad.png")
```
