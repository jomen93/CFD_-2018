```python
def f(x):
    return np.cos(x)-x/4.

def df(x):
    return -np.sin(x)-1/4.



def dx(f, x):
    return abs(0-f(x))
 
def newton_metod0(f, df, x0, e):
    delta = dx(f, x0)
    while delta > e:
        x0 = x0 - f(x0)/df(x0)
        delta = dx(f, x0)
    print 'Raiz en: ', x0
    print 'f(x) : ', f(x0)
    
    
newton_metod0(f,df,-3,1e-6)
newton_metod0(f,df,-2,1e-6)
newton_metod0(f,df,1,1e-6)
```
